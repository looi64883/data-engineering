````python
dtype = {
    "VendorID": "Int64",
    "passenger_count": "Int64",
    "trip_distance": "float64",
    "RatecodeID": "Int64",
    "store_and_fwd_flag": "string",
    "PULocationID": "Int64",
    "DOLocationID": "Int64",
    "payment_type": "Int64",
    "fare_amount": "float64",
    "extra": "float64",
    "mta_tax": "float64",
    "tip_amount": "float64",
    "tolls_amount": "float64",
    "improvement_surcharge": "float64",
    "total_amount": "float64",
    "congestion_surcharge": "float64"
}

parse_dates = [
    "tpep_pickup_datetime",
    "tpep_dropoff_datetime"
]

df = pd.read_csv(
    url,
    #nrows=100,
    dtype=dtype,
    parse_dates=parse_dates
)
````

``dtype = {...}``
This dictionary tells pandas exactly what data type each column should use when loading the CSV.

``parse_dates = [...]``
This list tells pandas:
“These columns are dates/timestamps, please parse them properly.”

``pd.read_csv(...)``
nrows=100                  Only read the first 100 rows
dtype=dtype	               Force specific data types for columns
parse_dates=parse_dates	   Convert listed columns into datetime objects

````python
# Batches
df_iter = pd.read_csv(
    url,
    dtype=dtype,
    parse_dates=parse_dates,
    iterator=True,
    chunksize=100000,
)
````

````python
!uv add tqdm
````

````python
from tqdm.auto import tqdm
````

tqdm wraps an iterable (like a loop) and visually shows progress
tqdm is a library used to show a progress bar for long-running operations (loops, downloads, data processing, etc.).

````python
# Iterate Over Chunks and Send into database
for df_chunk in tqdm(df_iter):
    df_chunk.to_sql(name='yellow_taxi_data', con=engine, if_exists='append')
````
